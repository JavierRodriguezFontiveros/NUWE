# Terraform

```python
import subprocess

# Comando para inicializar Terraform
init_command = ['terraform', 'init']
result = subprocess.run(init_command, capture_output=True, text=True)

print(result.stdout)  # Mostrar la salida
print(result.stderr)  # Mostrar los errores si los hubiera
```

```python
# Comando para crear un plan de Terraform
plan_command = ['terraform', 'plan']
result = subprocess.run(plan_command, capture_output=True, text=True)

print(result.stdout)  # Mostrar la salida del plan
print(result.stderr)  # Mostrar los errores si los hubiera
```

```python
import subprocess

# Comando para aplicar el plan de Terraform
apply_command = ['terraform', 'apply', '-auto-approve']
result = subprocess.run(apply_command, capture_output=True, text=True)

# Imprimir la salida estándar (stdout) y los errores (stderr)
print("Salida estándar:")
print(result.stdout)

# Mostrar errores si los hubiera
if result.stderr:
    print("\nErrores:")
    print(result.stderr)

# Verificar el código de salida para más información sobre el éxito o fallo
if result.returncode != 0:
    print(f"\nTerraform falló con código de salida: {result.returncode}")
else:
    print("\nTerraform se ejecutó con éxito.")
```

```python
import subprocess

# Ejecutar terraform destroy
try:
    # Llamar al comando terraform destroy
    subprocess.run(["terraform", "destroy", "-auto-approve"], check=True)
    print("Recursos eliminados correctamente.")
except subprocess.CalledProcessError as e:
    print(f"Error al ejecutar terraform destroy: {e}")
```
